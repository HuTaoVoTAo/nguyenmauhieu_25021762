```python
#W5A3
def prime(a):
    '''
        kiểm tra xem a có phải là số nguyên tố hay ko
    '''
    if a<2:
        return False
    else:
        for i in range (2,int(a**0.5)+1):
            if a%i==0:
                return False
        return True
n=int(input())
if prime(n):
    print("la so nguyen to")
else:
    print("ko phai la so nguyen to")
```

```python
#W5A4
def perfect(a):
    '''
        kiểm tra xem a có phải là số hoàn hảo hay ko
    '''
    sum=1
    for i in range (2,int(a**0.5)+1):
        if a%i==0:
            sum+=i
            if a//i!=i:
                sum+=a//i
    return sum==a
n=int(input())
if perfect(n):
    print("so hoan hao")
else:
    print("so ko hoan hao")
```

```python
#W5A5
def tim_gia_tri(a,x):
    '''
        tim kiem vi tri phan tu dau tien bang x trong list a
        neu ko tim duoc tra ve -1
    '''
    for i in range(0,len(a)):
        if a[i]==x:
            return i+1
    return -1
lst=[int(i) for i in input().split()]
n=int(input())
print(tim_gia_tri(lst,n))
```

```python
#W5A6
def giaithua(a):
    '''
        tinh giai thua cua a
    '''
    sum=1
    for i in range (1,a+1):
        sum*=i
    return sum
n=int(input())
print(giaithua(n))
```

```python
#W5A7
def pheptoan(a,b,op):
    '''
        kiem tra phep toan tu va thuc hien no voi a va b
    '''
    match op:
        case '-':
            return a-b
        case '+':
            return a+b
        case '*':
            return a*b
        case '/':
            return a^b
x,y,z=input().split()
x,y=float(x),float(y)
print(round((pheptoan(x,y,z))*100)/100)
```

```python
#W5A8
def rev_binary(a):
    '''
        chuyen doi tu thap phan sang nhi phan nguoc (retrun ra string)
    '''
    s=''
    if a==0:
        return '0'
    while a>0:
        if a%2:
            s+='1'
        else:
            s+='0'
        a=a//2
    return s
def hamming(a,b):
    '''
        kiem tra khoang cach hamming bang phep toan xor
    '''
    ans=0
    for i in range (0,len(b)):
        if a[i]!=b[i]:
            ans+=1
    for i in range (len(b),len(a)):
        if a[i]=='1':
            ans+=1
    return ans
x,y=map(int,input().split())
if x<y:
    x,y=y,x
print(hamming(rev_binary(x),rev_binary(y)))
```

```python
#W5A9
def tongchuso(a):
    '''
        tinh tong cac chu so
    '''
    sum=0
    while(a>0):
        sum+=a%10
        a//=10
    return sum
n=int(input())
print(tongchuso(n))
```
