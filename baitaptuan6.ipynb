```python
#W6A9
t=input().split()
dic1={t[i]:int(t[i+1]) for i in range(0,len(t)-1,2)}
t=input().split()
dic={t[i]:int(t[i+1]) for i in range(0,len(t)-1,2)}
#print(dic1,dic)
ans={}
for i in dic1:
    #print(i)
    if i in dic:
        ans[i]=dic[i]+dic1[i]
    else:
        ans[i]=dic1[i]
for i in dic:
    if i not in dic1:
        ans[i]=dic[i]
ans=dict(sorted(ans.items()))
print(ans)
```

```python
#W6A10
n=[int(i) for i in input().split()]
k=int(input())
d={}
c={}
for i in n:
    if i in d:
        d[i]+=1
        c[i]+=1
    else:
        d[i]=1
        c[i]=1
#print(d)
ans=[]
#print(n)
for i in n:
    #print(i)
    #print(i,' ',k-i)
    #print(d[i]+d[k-i])
    t=0
    if i==k-i:
        d[i]-=1
    while d[k-i]>t:
        ans.append((i,k-i))
        t+=1
    d[i]-=1
#print(ans)
for i in ans:
    print(i)
```

```python
#W6A12
n=[int(i) for i in input().split()]
n.sort()
dem=1
max=dem
for i in range (1,len(n)):
    if n[i]==n[i-1]:
        dem+=1
    else:
        dem=1
    if dem>max:
        ans=n[i]
        max=dem
print(ans)
    
```
